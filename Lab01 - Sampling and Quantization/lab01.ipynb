```python
#Nome: Leonardo Oliveira de Jesus ID: 12111EBI033
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_relogio = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

#plt.imshow(img, cmap='gray', vmin=0, vmax=255);

def reamostrar_imagem (imagem,DPI_original,dpi_escolhido):
    salto = int(DPI_original/dpi_escolhido) #DESAFIO: em numero real e arredondar na consulta (vetor _ fazer a media da janela de salto)
    return imagem[::salto,::salto]

DPI_relogio = 1250
dpis_escolhidos = [300,150,72]

img300 = reamostrar_imagem(img_relogio,DPI_relogio,300)
img150 = reamostrar_imagem(img_relogio,DPI_relogio,150)
img72  = reamostrar_imagem(img_relogio,DPI_relogio,72)

print(f"Dimensões 300 DPI: {img300.shape}")
print(f"Dimensões 150 DPI: {img150.shape}")
print(f"Dimensões 72  DPI: {img72.shape}")

# 4. Criar a visualização conjunta (1 linha e 4 colunas)
plt.figure(figsize=(20, 5)) # Define o tamanho da janela (largura, altura)

# Original
plt.subplot(1, 4, 1)
plt.imshow(img_relogio, cmap='gray')
plt.title(f"Original (1250 DPI)\n{img.shape}")
plt.axis('off')

# 300 DPI
plt.subplot(1, 4, 2)
plt.imshow(img300, cmap='gray')
plt.title(f"300 DPI\n{img300.shape}")
plt.axis('off')

# 150 DPI
plt.subplot(1, 4, 3)
plt.imshow(img150, cmap='gray')
plt.title(f"150 DPI\n{img150.shape}")
plt.axis('off')

# 72 DPI
plt.subplot(1, 4, 4)
plt.imshow(img72, cmap='gray')
plt.title(f"72 DPI\n{img72.shape}")
plt.axis('off')

plt.tight_layout() # Ajusta o espaçamento para não sobrepor títulos
plt.show()
```
